```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

x = np.arange(20)
y = np.random.randint(0,20,20)

x,y
plt.plot(x,y)


# y축을 20까지 보이게 하는법
plt.axis([0,20,0,20])
plt.yticks([0,5,10,15, 20])
plt.show()
```

```python
### 박스 그림 (box plot)
##수치형 데이터에 대한 정보 박스플랏이 가지고 있는 정보 (Q1,q2,q3,min,max)

plt.boxplot((x, y))
plt.title("box plot of y")
plt.show()

#엑스트라 : plot의 title을
```

```python
###막대그래프 ( bar plot)
#q\범주형 데이터의 값과 그 값의 크기를 직사각형으로 나타낸 그림
plt.bar(x,y)
plt.xticks(np.arange(0,20,1))
plt.show()

###xtics
```

```python
## 히스토그램
#도수분포를 직사각형의 막대 형태로 나타냄
##히스토그램과 막대그래프의 차이점
#"계급"으로 나타낸것이 특징 : 0,1,2가 아니라 0~2까지의 범주형데이터로 구성 후 그림을 그림

plt.hist(y, bins = np.arange(0,20,2))
plt.xticks(np.arange(0,20,2))
plt.show
```

```python
#파이그래프
z = [ 100,300,200,400]
plt.pie(z, labels=['1','2','3','4'])
plt.show()
```
