- 1969년 minsky와 papert의 perceptrons : 퍼셉트론의 한계를 지적하고 다층구조를 이용한 극복방안을 제시하였지만, 당기 기술로는 불가능.
- 1974년 webos는 오류역전파 알고리즘의 신경망 활용 가능성 확인(가능성만 확인)
- 1986년 Rumelhart, hinton에 의해 다층 퍼셉트론 이론 정립하고, 신경망 재부활
