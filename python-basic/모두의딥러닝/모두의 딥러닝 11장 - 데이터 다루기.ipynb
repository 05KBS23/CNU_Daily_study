# 2. 데이터 가공하기


많은 값을 단순히 나열하는 것은 눈에 들어오지도 않고 의미를 파악하기도 힘들다.
따라서 데이터를 잘 다루려면 데이터를 한 번더 가공해야함.
데이터를 가공할 때 우리가 무엇을 위해 작업을 하는지 목적을 잊어서는 안된다.
여기 데이터의 목적은 당뇨병발병을 예측하는 것이므로, 모든 정보는 당뇨병 발병과 어떤 관계가 있는지 중점에 놓아야 한다....


```python
print(df[['pregnant','class']].groupby(['pregnant'], as_index=False).mean().sort_values(by='pregnant', ascending=True))
```

-먼저 그룹바이 함수로 pregnant 정보를 기준으로 하는 새 그룹을 만든다,
-as_index=False는 pregnant 정보 옆에 새로운 인덱스를 만들어줌
-mean함수로 평균을 구하고, sort_values()함수로 pregnant 컬럼을 오름차순으로 정렬
-이를 출력하면 임신 횟수당 당뇨병 발병 확률을 구할 수 있다.

# 3. matplot을 이용해 그래프 표현하기

heatmap함수는 두 항목씩 짝을 지은 뒤 각각 어떤 패턴으로 변화하는지를 관찰하는 함수이다.  
두 항목이 전혀 다른 패턴으로 변화하고 있으면 0, 서로 비슷한 패턴으로 변할수록 1에 가까운 값을 출력한다.  

```python
# 데이터 간의 상관관계를 그래프로

colormap = plt.cm.gist_heat   #그래프 색상 정하기
plt.figure(figsize=(12,12))   #그래프 크기 정하기

# 그래프의 속성 결정. vmax의 값을 0.5로 지정해 0.5에 가까울 수록 밝은 색으로 표시
sns.heatmap(df.corr(),linewidths=0.1,vmax=0.5, cmap=colormap, linecolor='white', annot=True)
plt.show()
```

## 1. 함수, 라이브러리 호출 및 Seed 값 설정

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

import numpy
import tensorflow as tf


numpy.random.seed(3)
tf.random.set_seed(3)
```

## 랜덤 시드 부분이 이해가 잘 안 됬는데 책 146쪽에서 설명해주고 있다.

random함수를 써서 임의의 숫자를 만들어 내는 것처럼 보여도 이는 컴퓨토 안에 내장된 랜덤 테이블 중에서 하나를 불러내 표의 순서대로 숫자를 보여주는거
즉, seed 값을 설정한다는 것은 그 랜덤테이블중에서 몇 번째 테이블을 불러와 쓸지를 정하는 것이다.
따라서 seed 값이 같으면 똑같은 랜덤값을 출력한다.

넘파이 라이브러리를 사용하면서 텐서플로 기반으로 딥러닝을 구현할 때는 일정한 결괏값을 얻기 위해 넘파이 seed값과 텐서 seed 값을 같게 해야한다,.,

코드를 보면 numpy뿐만 아니라 tensorflow도 똑같은 seed로 설정해준 것을 알 수 있다.   
딥러닝을 구현할 때 일정한 결과값을 얻기 위한 것이며, 이렇게 똑같이 설정해주어도 출력값은 미세하게 다를 수 있다고 한다.   
이는 텐서플로우 자체 랜덤테이블 문제이므로 최종 딥러닝 결과를 여러 번 실행하여 평균을 구하는 것이 가장 적절하다


```python
# 데이터를 불러오기
dataset = numpy.loadtxt("/Users/KBS/Documents/study/python-basic/모두의딥러닝/pima-indians-diabetes.csv", delimiter=",")
X = dataset[:,0:8]
Y = dataset[:,8]
```

## 2. 모델 설정

10장에서는 은닉층이 총 2개지만 본 장에서는 총 3개로 진행보기로 한다.

```python

model = Sequential()
model.add(Dense(12, input_dim=8, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

```

# 3. 모델 컴파일


```python
model.compile(loss='binary_crossentropy',
             optimizer='adam',
             metrics=['accuracy'])
```



둘 중 하나를 결정하는 이항 분류이므로 오차함수는 바이너리 크로스엔트로피를 사용하고 최적화함수로 adam을 사용한다.

# 4. 모델 실행 및 결과 출력

```python
model.fit(X, Y, epochs=200, batch_size=10)

# 결과 출력
print("\n Accuracy: %.4f" % (model.evaluate(X, Y)[1]))
```
