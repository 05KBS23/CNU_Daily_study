# 1.환경 준비

- 기본 **라이브러리**와 대상 **데이터**를 가져와 이후 과정을 준비합니다.

```python
# 라이브러리 불러오기
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings(action='ignore')
%config InlineBackend.figure_format = 'retina'
```

```python
# 데이터 읽어오기
path = 'https://raw.githubusercontent.com/Jangrae/csv/master/income_happy.csv'
data = pd.read_csv(path)
```

```python
sns.scatterplot('income', 'happiness', data= data)
plt.show()
```

```python
# target 확인
target = 'happiness'

# 데이터 분리

```

```python
# 모듈 불러오기


# 7:3으로 분리
x = data.drop(target, axis = 1)
y = data.loc[:, target]
```

```python
from sklearn.model_selection import train_test_split

# 7:3으로 분리
x_train, x_test, y_train, y_test = train_test_split(x , y, 
                                                    test_size = 0.3,
                                                    shuffle = True, # 기본값 : 섞은 후 나눈 다
                                                    random_state=1) # 정수로 넣으면 건수, 실수로하면 퍼센트

# 4. 모델링
```

```python
# 1단계: 불러오기
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error

# 2단계: 선언하기 !         괄호를 반드시 넣어라.
model = LinearRegression()

# 3단계: 학습하기
model.fit(x_train,y_train)

# 4단계: 예측하기
y_pred = model.predict(x_test)
```

# 5.회귀 성능 평가

- 다양한 성능 지표로 회귀 모델 성능을 평가합니다.

**1) MAE(Mean Absolute Error)**

```python
# 모듈 불러오기

# 모듈 불러오기
from sklearn.metrics import mean_absolute_error

# 성능 평가
print('MAE:', mean_absolute_error(y_test,y_pred))
# 성능 평가
```

**2) MSE(Mean Squared Error)**

```python
# 모듈 불러오기

# 모듈 불러오기
from sklearn.metrics import mean_squared_error


# 성능 평가
print('MAE:', mean_squared_error(y_test,y_pred))
# 성능 평가
```

**3) RMSE(Root Mean Squared Error)**

```python
# 모듈 불러오기
# 모듈 불러오기

from sklearn.metrics import mean_squared_error

# 성능 평가
print('RMSE:', mean_squared_error(y_test,y_pred) ** 0.5) 
print('RMSE:', mean_squared_error(y_test,y_pred, squared=False))

# 성능 평가
```

**4) MAPE(Mean Absolute Percentage Error)**

```python
# 모듈 불러오기
# 모듈 불러오기
from sklearn.metrics import mean_absolute_percentage_error


# 성능 평가
print('MAPE:', mean_absolute_percentage_error(y_test,y_pred))


# 성능 평가
```

**5) R2-Score**

```python
# 모듈 불러오기
from sklearn.metrics import r2_score
print('R2:', r2_score(y_test,y_pred))


```
