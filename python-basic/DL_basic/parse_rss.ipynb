```python
urls = ['http://rss.etnews.com/Section901.xml',
        'http://rss.etnews.com/Section902.xml',
        'http://rss.etnews.com/Section903.xml',
        'http://rss.etnews.com/Section904.xml']
```

```python
def crawl_rss(urls):
  array_rss = []
  for url in urls:
    print("[Crawl RSS", url) # 확인
    parse_rss = feedparser.parse(url) 
    for p in parse_rss.entries: #기사를 하나씩 돌아다니면서 타이틀과 링크 따오기, n개에 대한 모든거 따오기
      array_rss.append({'title' : p.title, 'link' : p.link})
  return array_rss


def crawl_article(url, language = 'ko'): #url 다운
  print('[crawl Article]', url) #url 출력
  a = Article(url, language = language) # url을 집어너주고 a라는 변수에 넣어주면 본문 내용
  a.download()
  a.parse()
  a.parse
  return a.title, a.text

list_articles = crawl_rss(urls)

for article in list_articles: # 400개의 아티클
  _, text = crawl_article(article['link'])
  article['text'] = text

print(list_articles[0])
```
