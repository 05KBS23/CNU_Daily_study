### 1) CustomData yaml 생성

```python
import yaml
```

```python
cn_dict = {0:'person', 1:'vehicle', 2:'person_in_vehicle'}
```

```python
temp = {'path':'/content/datasets/',
        'train':'images/train/',
        'val':'images/val/',
        'nc':len(list(cn_dict)),
        'names':cn_dict}
```

```python
with open('/content/yolov5/data/custom.yaml' , 'w') as f:
    yaml.dump(temp, f)
```

### 4) train.py 실행

- 명령어 도움말 : python train.py -h

```python
!cd yolov5; python train.py -h
```

```python
!cd yolov5; python train.py \
    --data '/content/yolov5/data/street.yaml' \
    --cfg '/content/yolov5/models/yolov5s.yaml' \
    --weights '/content/yolov5/pretrained/yolov5s.pt' \
    --epochs 1000 \
    --patience 3 \
    --img 640 \
    --project 'trained' \
    --name 'train_street' \
    --exist-ok
    # --device cpu
```

### 6) detect.py 실행

- 명령어 도움말 : python detect.py -h

```python
!cd yolov5; python detect.py \
    --weights '/content/yolov5/trained/train_street/weights/best.pt' \
    --source '/content/yolov5/data/images/' \
    --project '/content/yolov5/detected' \
    --name 'images' \
    --img 640 \
    --conf-thres 0.5 \
    --iou-thres 0.5 \
    --line-thickness 2 \
    --exist-ok 
    # --device CPU
```
