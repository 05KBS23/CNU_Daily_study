#### [실습문제3] 데이터 확인 및 처리

* head 와 tail 을 보았을때, 어느 데이터만 가져와야 할지 생각 해 봅시다.

```python
seoul_people = pd.read_csv('rawdata/1.0 주민등록 데이터.csv', sep="\t", encoding = "cp949")
seoul_people
```

```python
seoul_people['소계'].sum()
```

```python
seoul_people = seoul_people[['동별(2)', '소계', '남자', '여자', '65세이상고령인']]
seoul_people = seoul_people.drop(0)
seoul_people
```

```python
seoul_old_people = seoul_people[['동별(2)', '소계', '남자', '여자', '65세이상고령인']]
```

```python
seoul_old_people['65세이하'] = seoul_old_people['소계'] - seoul_old_people['65세이상고령인']
```

```python
seoul_old_people['청년 비율'] = seoul_old_people['65세이하'] / seoul_old_people['소계']
```

```python
seoul_old_people['노인 비율'] = seoul_old_people['65세이상고령인'] / seoul_old_people['소계']
seoul_old_people
```
